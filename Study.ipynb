```python
import pandas as pd

file_one = "census_data.csv"

file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
file_one_df.head()
```

```python
file_one_df = file_one_df.rename(columns={'Name':'NAME'})
file_one_df.head()
```

```python
file_two = "ethnics_data.csv"

file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1")
file_two_df.head()
```

```python
file_two_df.columns
```

```python
file_two_df = file_two_df.drop(columns=['White','Black','Asian', 'Latino','state'])
```

```python
file_two_df.head()
```

```python
census_study = file_two_df.merge(file_one_df, how='outer', on='NAME')
```

```python
census_study.head()
```

```python
census_study = census_study.rename(columns={'NAME':'state_name','Total':'Total state ethnicity'})
```

```python
census_study.head()
```

```python
census_study.to_csv("census_study.csv", encoding="utf-8", index=False)
```
