```python
import pandas as pd
import requests
import numpy as np
import json

# more documentation https://github.com/datamade/census/blob/master/census/tests/test_census.py

#Using pip install census 
#Using pip instal us 
from census import Census
from us import states

from config import api_key
c = Census(api_key, year=2010)
```

```python
ethnics_data = c.sf1.get(
            ("NAME",
             'P008001', # total
             'P008003', # white
             'P008004', # black
             'P008006', # asian
             'P008010', # latino
            ),
            geo={'for': 'state:*'},
        )
ethnics_data
```

```python
ethnics_data = pd.DataFrame(ethnics_data)
ethnics_data
```

```python
ethnics_data = ethnics_data.rename(columns={'P008001':'Total',
                                            'P008003':'White',
                                            'P008004':'Black',
                                            'P008006':'Asian',
                                            'P008010':'Latino'})
ethnics_data
```

```python
ethnics_data['White Avg'] = ethnics_data['White']/ethnics_data['Total'] * 100
ethnics_data['Black Avg'] = ethnics_data['Black']/ethnics_data['Total'] * 100
ethnics_data['Asian Avg'] = ethnics_data['Asian']/ethnics_data['Total'] * 100
ethnics_data['Latino Avg'] = ethnics_data['Latino']/ethnics_data['Total'] * 100
ethnics_data
```

```python
ethnics_data.to_csv("ethnics_data.csv", encoding="utf-8", index=False)
```
